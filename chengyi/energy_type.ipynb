```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
baltic_energy = pd.read_csv('/home/ctang/ttmp/Datathon/data_and_starter_code/data/baltic_energy_type.csv')
```

```python
baltic_energy.head()
```

```python
# sum up all generation by Area and Production Type by year
# find year by converting MTU (CET/CEST) to string and take the 6th to 9th character
baltic_energy['Year'] = baltic_energy['MTU (CET/CEST)'].astype(str).str[6:10]
baltic_energy['Year'] = baltic_energy['Year'].astype(int)
baltic_energy['Generation (MW)'] = baltic_energy['Generation (MW)'].astype(float)
```

```python
# sum up all generation by Area and Production Type by year
baltic_energ_sum = baltic_energy.groupby(['Year', 'Area', 'Production Type'])['Generation (MW)'].sum().reset_index()
```

```python
# graph pie chart of Estonia production types
estonia = baltic_energ_sum[baltic_energ_sum['Area'] == 'Estonia']
estonia_2024 = estonia[estonia['Year'] == 2024]
estonia_2021 = estonia[estonia['Year'] == 2021]

# sort by generation
estonia_2024 = estonia_2024.sort_values(by='Generation (MW)', ascending=False)
estonia_2021 = estonia_2021.sort_values(by='Generation (MW)', ascending=False)

# only plot the ones with 0.5% or more, others are grouped into 'Other'
estonia_2024 = estonia_2024[estonia_2024['Generation (MW)'] > 0.01 * estonia_2024['Generation (MW)'].sum()]
estonia_2021 = estonia_2021[estonia_2021['Generation (MW)'] > 0.01 * estonia_2021['Generation (MW)'].sum()]
```

```python
estonia_2024
```

```python
# Define a color palette for consistency
colors = plt.cm.Set3.colors  # or try tab20c, Pastel1, etc.

fig, ax = plt.subplots(1, 2, figsize=(14, 7), dpi=100)

# Pie chart for 2021
wedges1, texts1, autotexts1 = ax[0].pie(
    estonia_2021['Generation (MW)'],
    labels=estonia_2021['Production Type'],
    autopct='%1.1f%%',
    startangle=90,
    colors=colors,
    wedgeprops=dict(edgecolor='black')
)
ax[0].set_title('Estonia 2021', fontsize=14, weight='bold')

# Pie chart for 2024
wedges2, texts2, autotexts2 = ax[1].pie(
    estonia_2024['Generation (MW)'],
    labels=estonia_2024['Production Type'],
    autopct='%1.1f%%',
    startangle=90,
    colors=colors,
    wedgeprops=dict(edgecolor='black')
)
ax[1].set_title('Estonia 2024', fontsize=14, weight='bold')

# Make all text consistent and readable
for autotext in autotexts1 + autotexts2:
    autotext.set_color('black')
    autotext.set_fontsize(10)

for text in texts1 + texts2:
    text.set_fontsize(10)

# Set a main title
plt.suptitle('Estonia Electricity Generation by Production Type: 2021 vs 2024', fontsize=16, weight='bold')

# Adjust layout to prevent overlap
plt.tight_layout(rect=[0, 0, 1, 0.95])

plt.show()
```

```python
latvia = baltic_energ_sum[baltic_energ_sum['Area'] == 'Latvia']
latvia_2024 = latvia[latvia['Year'] == 2024]
latvia_2021 = latvia[latvia['Year'] == 2021]
latvia_2024 = latvia_2024.sort_values(by='Generation (MW)', ascending=False)
latvia_2021 = latvia_2021.sort_values(by='Generation (MW)', ascending=False)
latvia_2024 = latvia_2024[latvia_2024['Generation (MW)'] > 0.01 * latvia_2024['Generation (MW)'].sum()]
latvia_2021 = latvia_2021[latvia_2021['Generation (MW)'] > 0.01 * latvia_2021['Generation (MW)'].sum()]
fig, ax = plt.subplots(1, 2, figsize=(14, 7), dpi=100)
wedges1, texts1, autotexts1 = ax[0].pie(
    latvia_2021['Generation (MW)'],
    labels=latvia_2021['Production Type'],
    autopct='%1.1f%%',
    startangle=90,
    colors=colors,
    wedgeprops=dict(edgecolor='black')
)
ax[0].set_title('Latvia 2021', fontsize=14, weight='bold')
wedges2, texts2, autotexts2 = ax[1].pie(
    latvia_2024['Generation (MW)'],
    labels=latvia_2024['Production Type'],
    autopct='%1.1f%%',
    startangle=90,
    colors=colors,
    wedgeprops=dict(edgecolor='black')
)
ax[1].set_title('Latvia 2024', fontsize=14, weight='bold')
for autotext in autotexts1 + autotexts2:
    autotext.set_color('black')
    autotext.set_fontsize(10)
for text in texts1 + texts2:
    text.set_fontsize(10)
plt.suptitle('Latvia Electricity Generation by Production Type: 2021 vs 2024', fontsize=16, weight='bold')
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
lithuania = baltic_energ_sum[baltic_energ_sum['Area'] == 'Lithuania']
lithuania_2024 = lithuania[lithuania['Year'] == 2024]
lithuania_2021 = lithuania[lithuania['Year'] == 2021]
lithuania_2024 = lithuania_2024.sort_values(by='Generation (MW)', ascending=False)
lithuania_2021 = lithuania_2021.sort_values(by='Generation (MW)', ascending=False)
lithuania_2024 = lithuania_2024[lithuania_2024['Generation (MW)'] > 0.01 * lithuania_2024['Generation (MW)'].sum()]
lithuania_2021 = lithuania_2021[lithuania_2021['Generation (MW)'] > 0.01 * lithuania_2021['Generation (MW)'].sum()]
fig, ax = plt.subplots(1, 2, figsize=(14, 7), dpi=100)
wedges1, texts1, autotexts1 = ax[0].pie(
    lithuania_2021['Generation (MW)'],
    labels=lithuania_2021['Production Type'],
    autopct='%1.1f%%',
    startangle=90,
    colors=colors,
    wedgeprops=dict(edgecolor='black')
)
ax[0].set_title('Lithuania 2021', fontsize=14, weight='bold')
wedges2, texts2, autotexts2 = ax[1].pie(
    lithuania_2024['Generation (MW)'],
    labels=lithuania_2024['Production Type'],
    autopct='%1.1f%%',
    startangle=90,
    colors=colors,
    wedgeprops=dict(edgecolor='black')
)
ax[1].set_title('Lithuania 2024', fontsize=14, weight='bold')
for autotext in autotexts1 + autotexts2:
    autotext.set_color('black')
    autotext.set_fontsize(10)
for text in texts1 + texts2:
    text.set_fontsize(10)
plt.suptitle('Lithuania Electricity Generation by Production Type: 2021 vs 2024', fontsize=16, weight='bold')
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
import matplotlib.pyplot as plt

# Define consistent color mapping
prod_types = sorted(set(estonia_2021['Production Type']) |
                    set(latvia_2021['Production Type']) |
                    set(lithuania_2021['Production Type']))

# Assign a color to each production type
color_map = {ptype: plt.cm.tab20(i) for i, ptype in enumerate(prod_types)}
colors_est = [color_map[ptype] for ptype in estonia_2021['Production Type']]
colors_lat = [color_map[ptype] for ptype in latvia_2021['Production Type']]
colors_lit = [color_map[ptype] for ptype in lithuania_2021['Production Type']]

# Create subplots
fig, ax = plt.subplots(1, 3, figsize=(21, 7), dpi=100)

# Estonia
wedges1, _, autotexts1 = ax[0].pie(
    estonia_2021['Generation (MW)'],
    autopct='%1.1f%%',
    startangle=90,
    colors=colors_est,
    wedgeprops=dict(edgecolor='black')
)
ax[0].set_title('Estonia 2021', fontsize=14, weight='bold')

# Latvia
wedges2, _, autotexts2 = ax[1].pie(
    latvia_2021['Generation (MW)'],
    autopct='%1.1f%%',
    startangle=90,
    colors=colors_lat,
    wedgeprops=dict(edgecolor='black')
)
ax[1].set_title('Latvia 2021', fontsize=14, weight='bold')

# Lithuania
wedges3, _, autotexts3 = ax[2].pie(
    lithuania_2021['Generation (MW)'],
    autopct='%1.1f%%',
    startangle=90,
    colors=colors_lit,
    wedgeprops=dict(edgecolor='black')
)
ax[2].set_title('Lithuania 2021', fontsize=14, weight='bold')

# Format text
for autotext in autotexts1 + autotexts2 + autotexts3:
    autotext.set_color('black')
    autotext.set_fontsize(10)

# Create a shared legend
handles = [plt.Line2D([0], [0], marker='o', color='w', label=ptype,
                      markerfacecolor=color_map[ptype], markersize=10)
           for ptype in prod_types]
fig.legend(handles=handles, loc='center right', title='Production Type', fontsize=18, title_fontsize=12, 

plt.suptitle('Baltic States Electricity Generation by Production Type: 2021', fontsize=16, weight='bold')
plt.tight_layout(rect=[0, 0, 0.9, 0.95])  # Make room for legend on the right
plt.show()
```
